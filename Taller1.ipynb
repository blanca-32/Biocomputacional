```python
# Blanca Cerinza
# titulo: Conteo de polinizadores en plantas endemicas de Colombia
# Injtroducción: Realizamos esta investigacion sobre los polinizadores con platnas endemicas de colombia, fuimos a diferentes lugares de colombia, nuestro objetivo es ver cuantos polinizadores se acercan a ellas. Con el fin de entender porque algunas tienen mayor gusto a unas plantas en especifico. Para a posteriori hacer un analiss quimico y entener cuales son los compuestos atractivos.
# Descripción del procedimiento: Lo quisimos hacer en en un pesacio reducido a una parcela de 50mt, por temas de tiempo y costos. Por ende vemos especies con muy pocos conteos. Logrando centralizar la informacion en que la mayor cantidad de especies son las mas comunes como melastomataceae y asteraceae. Haciendo observaciones por 3 dias. Siendo mayor en las que tienen flores mas llamativas.
!pip install pandas matplotlib --quiet #instalar la librería pandas

# Importamos las librerías que vamos a utilizar
import pandas as pd
import matplotlib.pyplot as plt

# Aquí usamos un diccionario para guardar los datos del conteo de polen por familia
data = {
    'Familia': ['Melastomataceae Miconia', 'Orchidaceae Cattleya', 'Bromeliaceae Guzmania', 'Asteraceae Espeletia',
                'Ericaceae Cavendishia', 'Lauraceae Ocotea', 'Rubiaceae Palicourea', 'Gesneriaceae Columnea',
                'Araceae Anthurium', 'Piperaceae Piper', 'Euphorbiaceae Croton', 'Sapindaceae Paullinia',
                'Fabaceae Inga', 'Campanulaceae Centropogon', 'Malvaceae Theobroma'],
    'Conteo': [102, 30, 22, 90, 22, 41, 60, 87, 58, 36, 25, 34, 43, 3, 18]
}

# Convertimos el diccionario en un DataFrame (tabla de datos) usando pandas
df = pd.DataFrame(data)

# Calculamos el porcentaje que representa cada tipo de polen respecto al total
df['Porcentaje'] = df['Conteo'] / df['Conteo'].sum() * 100

# Ordenamos el DataFrame de mayor a menor según el porcentaje y lo guardamos
df_ordenado = df.sort_values(by='Porcentaje', ascending=False)

# Mostramos la tabla ordenada
print(df_ordenado)

# Gráfica ordenada
plt.figure(figsize=(10,6))
plt.barh(df_ordenado['Familia'], df_ordenado['Porcentaje'], color='skyblue')
plt.xlabel('% de polen')
plt.title('Distribución porcentual del polen por familia')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
